```python
import matplotlib
from sklearn.manifold import TSNE
from kaldi_to_numpy import ark2dict 
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
ivec_path = "/macierz/home/s174520/ivector-xvector/ivector/data/feat/ivectors_enroll_mfcc/ivector.1.ark"
ivec_dict = ark2dict(ivec_path)
print len(ivec_dict)
print(ivec_dict.keys())
```

```python
actors, ivecs_raw = zip(*ivec_dict.items())
ivecs = np.array(ivecs_raw)
print actors
```

```python
ivecs_emb = TSNE(n_components=2).fit_transform(ivecs)
```

```python
x, y = ivecs_emb[:, 0], ivecs_emb[:, 1] 
```

```python
plt.scatter(x, y)
for actor_name, actor_x, actor_y in zip(ivec_dict.keys(), x, y):
    plt.annotate(
        actor_name,
        xy=(actor_x, actor_y),
        arrowprops=dict(facecolor='red', width=2.0))
plt.axes().set_aspect('equal')
plt.show()
```
